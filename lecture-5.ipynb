```python

from sklearn import datasets
from sklearn.metrics import confusion_matrix

iris = datasets.load_iris()
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test,Y_train, Y_test = train_test_split(iris.data, iris.target,test_size=0.2)






```

```python
clf = LinearRegression()
clf.fit(X_train,Y_train)
Y_pred = clf.predict(X_test).astype(int)
#Y_pred
print(confusion_matrix(Y_test,Y_pred))      #to predict which value is going from correct classifier to which wrong classifier
Y_pred.sort()   #to check whether predicted is whether a row or column
print(Y_pred)
Y_test.sort()
print(Y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,Y_pred))
```

```python
from sklearn.model_selection import cross_val_score
clf2 = LinearRegression()
cross_val_score(clf2,iris.data, iris.target,cv=5)
```
